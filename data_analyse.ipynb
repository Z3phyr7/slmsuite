```python
x1 = np.array([spot['x0'] for spot in spot1])
y1 = np.array([spot['y0'] for spot in spot1])
x2 = np.array([spot['x0'] for spot in spot2])
y2 = np.array([spot['y0'] for spot in spot2])

taus1, adev1, _, _ = allantools.adev(x1, data_type='freq')
taus2, adev2, _, _ = allantools.adev(x2, data_type='freq')
taus3, adev3, _, _ = allantools.adev(y1, data_type='freq')
taus4, adev4, _, _ = allantools.adev(y2, data_type='freq')

fig,axis = plt.subplots()
axis.plot(taus1, adev1, label='x of spot1', color='red')
axis.plot(taus2, adev2, label='x of spot2', color='blue')
axis.plot(taus3, adev3, label='y of spot1', color='green')
axis.plot(taus4, adev4, label='y of spot2', color='yellow')

axis.set_xscale('log')
axis.set_yscale('log')
axis.set_xlabel('tau')
axis.set_ylabel('ADEV')
axis.set_title('Allan Deviation of x and y of spot1 and spot2')
axis.legend()
plt.show()
```
