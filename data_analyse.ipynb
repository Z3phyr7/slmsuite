```python
gaussian_params = read_gaussian_params_from_file(filename='gaussian_params_expo_3e-3.txt')
spot1,spot2 = sort_spot(gaussian_params=gaussian_params, x1=3089,x2=2493)
deviation1 = deviation_anlys(spot_list=spot1)
deviation2 = deviation_anlys(spot_list=spot2)
print("deviation of 1 : {}".format(deviation1))
print("deviation of 2 : {}".format(deviation2))
```

```python
print("deviation of 1 : {}".format(sum(deviation1)/len(deviation1)))
print("deviation of 2 : {}".format(sum(deviation2)/len(deviation2)))
```

```python
x1 = np.array([element['x0'] for element in spot1])
y1 = np.array([element['y0'] for element in spot1])
x2 = np.array([element['x0'] for element in spot2])
y2 = np.array([element['y0'] for element in spot2])

x1_avr = np.mean(x1)
y1_avr = np.mean(y1)
x2_avr = np.mean(x2)
y2_avr = np.mean(y2)

spread1 = np.mean(np.sqrt((x1-x1_avr)**2 + (y1-y1_avr)**2))
spread2 = np.mean(np.sqrt((x2-x2_avr)**2 + (y2-y2_avr)**2))
```

```python
print("spread 1: {}".format(spread1))
print("spread 2: {}".format(spread2))
```
