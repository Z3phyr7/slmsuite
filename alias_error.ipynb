```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = np.linspace(0, 4.8, 500)
y = 10.231482 * r**2
```

```python
y_SLM = np.array(y[i] for i in range(len(r)) if i % 25 == 0)
```

```python
fig, ax = plt.subplots()
ax.plot(r, y, label='Theoretical phase pattern')
ax.plot(r, y_SLM, 'o', label='Real SLM Pattern')
ax.set(xlabel='r', ylabel='y', title='Phase pattern')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f(x):
    return x**2  # Example function: f(x) = x^2

# Define the limits of integration
a = 0
b = 5
pi = 3.14159265359
# Choose the number of rectangles
num_rectangles = 25

# Calculate the width of each rectangle
dx = (b - a) / num_rectangles

# Calculate the heights of the rectangles
x_values = np.linspace(a, b, num_rectangles + 1)
y_values = f(x_values[:-1])  # Evaluate the function at the left endpoints of each rectangle

# Plot the rectangles
plt.bar(x_values[:-1], y_values, width=dx, align='edge', color='blue', edgecolor='black', label = 'Real SLM Pattern')

# Plot the function
x = np.linspace(a, b, 100)
plt.plot(x, f(x), color='red', label = 'Theoretical phase pattern')

plt.axhline(y=2*pi, color='green', linestyle='--')
plt.axhline(y=4*pi, color='green', linestyle='--')
plt.axhline(y=6*pi, color='green', linestyle='--')


# Add labels and title
plt.xlabel('radius r')
plt.ylabel('phase pattern f(r)')
plt.title('Phase pattern of a SLM')

# Show plot
# plt.grid(True)
plt.legend()
plt.show()
```
